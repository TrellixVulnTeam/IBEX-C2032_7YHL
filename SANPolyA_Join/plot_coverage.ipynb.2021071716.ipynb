```r
plotPASCoverage <- function(pas,data,start,end){
    
    if(is.numeric(pas)){
        df =  data.frame(position = (-(end-start)/2):((end-start)/2),RPM = as.vector(as.matrix(data[pas,start:end])))
    }
    else{
        cov = subset(data,data$V1 == pas)[,start:end]
        df = data.frame(position = (-(end-start)/2):((end-start)/2),RPM = as.vector(as.matrix(cov)))
    }
    print(pas)
    p = ggplot(data=df,aes(x=position,y=RPM))+
    geom_line()+
    labs(title=paste("Coverage of PAS ",pas,sep=''))+
    scale_x_continuous(breaks = seq(-(end-start)/2, (end-start)/2, by = (end-start)/8))+
    theme_bw()+
    theme(axis.title=element_text(size=20),
          axis.text=element_text(size=20,vjust=0.5,hjust=0.5,angle=0),
          legend.title=element_blank(),
          plot.title = element_text(size=24,face = "bold",hjust = 0.5),
          legend.text=element_text(size=18,color="black",face="bold"))
    print(p)
}
```

```r
data= read.table('tianbin_data/bl6.pAs.compare2.txt',header=TRUE)
sp <- ggscatter(data, x='exp_usage',y='predict_usage',
                 shape = 19,size=1,
                 add = "reg.line",  # Add regressin line
                 add.params = list(color = "blue", fill = "lightgray"), # Customize reg. line
                 conf.int = TRUE, # Add confidence interval
                )
sp + stat_cor(method = "spearman", cor.coef.name = "rho",label.x = 0.45, label.y = 1.02)
#sp + stat_cor(method = "pearson",label.x = 0.45, label.y = 1.02)
```
